```python
import os
currentpath = os.getcwd()

def reportGen(info,img_path,data,data1):
    img_list =img_path[0]
    img_list=os.listdir(img_list)

    html_string='<html><head><title></title></head><body><table>'
    html_string +="<tr>"
    for head in range(0,len(info)):
        html_string +="<th width='40%'>"+ info[head] +"</th>"
    html_string +="</tr>"

    for sl in range(0, len(img_list)):
        html_string +="<tr>"
        for path in range(0, len(img_path)):
            ld=os.listdir(img_path[path])
            html_string +="<td>"+ ld[sl] +"<br/><img width='300' src='"+ img_path[path] + ld[sl] +"'></td>"
        html_string +="<td>"+ str(data[sl]) +"</td>"
        html_string +="<td>"+ str(data1[sl]) +"</td>"
        html_string +="</tr>"

    html_string +='</table></body></html>'
    f = open('Results/report.html','w')
    f.write(html_string)
    f.close()
    
tablehead=['original image','Hard Exudate','Precision','Recall']
report_image_path=os.path.join(currentpath,'data','images','Training')+'/'
report_image_path1=os.path.join(currentpath,'Results','HardExodus','Training')+'/'
filepath=[report_image_path, report_image_path1]
k=[f"{i}" for i in range(len(os.listdir(report_image_path1)))]
reportGen(tablehead,filepath,k,k)
```
