```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Image Path
image_path = "data/images/test/"
images = os.listdir(image_path)
images.sort()
print(image_path + "IDRiD_65.jpg" )
img= cv2.imread(image_path + "IDRiD_55.jpg",cv2.IMREAD_COLOR)
print("Original Image")
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
imshow(img)
```

```python
# convert to green channel 
B,G,R=cv2.split(img)
imshow(G)

kernel = np.ones((11,11),np.uint8)
img_g = G

# top hat 
tophat_img=cv2.morphologyEx(img_g,cv2.MORPH_TOPHAT,kernel)
# tophat_img=cv2.morphologyEx(tophat_img,cv2.MORPH_TOPHAT,kernel)
imshow(tophat_img)

# bottom hat
blackhat = cv2.morphologyEx(tophat_img, cv2.MORPH_BLACKHAT, kernel)
print(" Image")
imshow(blackhat)
```

```python
kernel = np.ones((11,11),np.uint8)
dilation = cv2.dilate(img,kernel,iterations = 1)
imshow(dilation)
tophat = cv2.morphologyEx(img, cv2.MORPH_TOPHAT, kernel)
print(" Image")
imshow(tophat)
```
