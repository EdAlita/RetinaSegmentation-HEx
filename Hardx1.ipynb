```python
# Ref: https://www.mdpi.com/1424-8220/19/3/695
#Import package
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np
```

##Funcitions 

```python
#------------#
# Function for Visualising Images
def imshow(img):
    plt.axis('off')
    plt.imshow(img, 'gray')
    plt.show()
```

```python
#Input image
image_path = "data/images/test/"
image_path = "data/images/training/"


# images = os.listdir(image_path)
# images.sort()
# print(image_path + "IDRiD_10.jpg" )
img= cv2.imread(image_path + "IDRiD_10.jpg",cv2.IMREAD_GRAYSCALE)
print("Original Image")

cv2.imshow('Img',img)
```

```python
img_hist=cv2.equalizeHist(img)
imshow(img_hist)
```

```python
img_filtered = cv2.medianBlur(img_hist, 3)
imshow(img_filtered)
```

```python
gamma = 1
img_gamma = np.power(img_filtered/255.0, gamma)
img_gamma = np.uint8(img_gamma*255)

imshow(img_gamma)
```
